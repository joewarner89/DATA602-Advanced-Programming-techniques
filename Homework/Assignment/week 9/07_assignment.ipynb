Cheking the average houses that were sold help us understand a few things that buyers are actually after in the market.

```python
average_price = house['SALE_PRC'].mean()
# Plot
plt.figure(figsize=(8, 6))
plt.hist(house['SALE_PRC'], bins=30, color='skyblue', edgecolor='black')
plt.axvline(x=average_price, color='red', linestyle='--', label=f'Average Price: ${average_price:.2f}')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Housing Prices in Miami')
plt.legend()
plt.show()

```

we are going to change 3 columns names to better identify them. 
* SALE_PRC to price
* TOT_LVG_AREA to sqt_foot
* age to year_built

```python
# Rename a column
house.rename(columns={'SALE_PRC': 'price'}, inplace=True)
house.rename(columns={'TOT_LVG_AREA': 'sqt_foot'}, inplace=True)
house.rename(columns={'age': 'year_built'}, inplace=True)

print(house.columns.to_list())
```

It seems like some of the features of the dataset has been properly cleaned for data analysis. so we do not have missing data and incorrect. we are going to create new column school district grade based on how much the house was sold. if a house is sold less than 200k then the school district categorie would be labelled as 'bad' or more than 399k 

```python
# Define function to assign labels based on price
def label_school_district(price):
    if price < 300000:
        return 'Bad'
    elif price >= 300000 and price < average_price:
        return 'Average'
    elif price >= average_price and price < 500000:
        return 'Good'
    else:
        return 'Excellent'

# Create new column 'School District'
house['School District'] = house['price'].apply(label_school_district)

# Display the first few rows to verify the result
print(house.head())
```

Droping columns we don't need during our analysis 

```python

# Drop multiple columns
house.drop(['avno60plus', 'LND_SQFOOT'], axis=1, inplace=True)
print(house.columns.to_list())
```

dropping multiple rows in the dataset

```python
# Drop multiple rows by index
print(house.shape)
house.drop(index=[1, 2, 3], inplace=True)
print(house.shape)
```

```python
# Sort data based on multiple variables
house_sorted = house.sort_values(by=['price', 'OCEAN_DIST'], ascending=[True, False])
house_sorted.head(10)
```

we are going to select the neigbohood with best school district where houses sold for more than $ 900k

```python
import folium
miami_map = house[house['price'] > 900000]
# Create a map centered around Miami
miami_map = folium.Map(location=[25.7617, -80.1918], zoom_start=12)

# Sample latitude and longitude data
latitude = house['LONGITUDE']  # Example latitude values
longitude = house['LATITUDE']  # Example longitude values

# Add markers for each latitude and longitude point
for lat, lon in zip(latitude, longitude):
    
    folium.Marker(location=[lat, lon]).add_to(miami_map)

# Add feature to recognize location on click
folium.ClickForMarker(popup='Click to add a marker').add_to(miami_map)

# Save the map to an HTML file
miami_map.save("miami_map.html")

# Display the map
miami_map
```

Group your dataset by one column, and get the mean, min, and max values by group



```python
house_size = house.groupby('sqt_foot').agg({'price': ['mean','min', 'max']})
# Rename the columns for clarity
house_size.columns = ['mean_value', 'min_value', 'max_value']

# Display the grouped data
print(house_size)
```
